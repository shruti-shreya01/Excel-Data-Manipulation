# Melt the second dataframe

```python
import pandas as pd

# Assuming `df` is your original dataframe
# Melting the dataframe
melted_df = pd.melt(d2, id_vars=['timestamp'], var_name='location_measurement', value_name='Value')

# Extract location and measurement columns from 'location_measurement'
melted_df['location'] = melted_df['location_measurement'].apply(lambda x: x.split()[0])
melted_df['Measurement'] = melted_df['location_measurement'].apply(lambda x: " ".join(x.split()[1:]))

# Drop the 'location_measurement' column
melted_df.drop(columns=['location_measurement'], inplace=True)

# Format the timestamp column
melted_df['timestamp'] = pd.to_datetime(melted_df['timestamp']).dt.strftime('%m/%d/%Y')

# Reorder columns to match the desired output
final_df = melted_df[['timestamp', 'location', 'Measurement', 'Value']]

# Display the result
final_df.head()
```

## sample working

```python
import pandas as pd

# Sample DataFrame with additional columns
data = {
    "Category": ["A", "A", "A", "B", "B", "C", "C", "C"],
    "Value": [1, 2, 3, 4, 5, 6, 7, 8],
    "Column1": ["X", "Y", "Z", "P", "Q", "R", "S", "T"],
    "Column2": [10, 20, 30, 40, 50, 60, 70, 80]
}

df = pd.DataFrame(data)

# Add a helper column to create a unique index for each category
df["Index"] = df.groupby("Category").cumcount()

# Pivot the DataFrame for only the Category and Value columns
pivot_df = df[["Category", "Value", "Index"]].pivot(index="Index", columns="Category", values="Value")

# Reset index for a cleaner display (optional)
pivot_df.reset_index(drop=True, inplace=True)

# Merge the pivoted DataFrame with the original columns (excluding Category and Value)
final_df = pd.merge(df.drop(columns=["Category", "Value"]), pivot_df, left_on="Index", right_index=True)

# Drop the 'Index' column if it's no longer needed
final_df.drop(columns=["Index"], inplace=True)

print("Final DataFrame:")
print(final_df)
```
